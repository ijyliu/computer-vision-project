# Create Selected Images

Limit classes, get rid of low resolution images

Create new train-test split in folders

Create new, filtered file with class labels

```python
# Packages
import pandas as pd
import openpyxl
import os
import shutil
import time
import numpy as np
```

```python
# Load "~/Box/INFO 290T Project/Intermediate Data/cars_annos.xlsx"
cars_annos = pd.read_excel("~/Box/INFO 290T Project/Intermediate Data/cars_annos.xlsx")
cars_annos
```

## Limit Images

```python
limited_cars_annos = cars_annos.query('Class in ["SUV", "Sedan", "Convertible", "Pickup"] and width > 350 and height > 200').reset_index(drop=True)
print(limited_cars_annos['Class'].value_counts())
print(limited_cars_annos[['width', 'height']].describe())
limited_cars_annos
```

```python
# Variable for originally train or test
limited_cars_annos['orig_train_or_test'] = np.where(limited_cars_annos['test'] == 0, 'train', 'test')
limited_cars_annos['orig_train_or_test'].value_counts()
```

## Create New Train-Test Split in Folders

```python
# Print first value of im_path
print(limited_cars_annos['im_path'][0])
```

```python
# Iterate over rows
# Copy image to "~/Box\INFO 290T Project\Intermediate Data\Selected Images\train" (or test) with Class and train (or test) appended to front of file name
# List of new file paths
new_file_paths = []
new_file_names = []
# Clear directories
start = time.time()
shutil.rmtree(os.path.expanduser('~') + "/Box/INFO 290T Project/Intermediate Data/Selected Images/train")
shutil.rmtree(os.path.expanduser('~') + "/Box/INFO 290T Project/Intermediate Data/Selected Images/test")
os.makedirs(os.path.expanduser('~') + "/Box/INFO 290T Project/Intermediate Data/Selected Images/train")
os.makedirs(os.path.expanduser('~') + "/Box/INFO 290T Project/Intermediate Data/Selected Images/test")
end = time.time()
print("Time to clear directories in min: ", (end - start) / 60)
# start timer
start = time.time()
for start_im_path, start_filename, test_80_20_val, Class, orig_train_or_test in limited_cars_annos[['im_path', 'filename', 'test_80_20', 'Class', 'orig_train_or_test']].itertuples(index=False):
    if test_80_20_val == 0:
        source = os.path.expanduser('~') + start_im_path
        new_file_name = Class + "_train_orig_" + orig_train_or_test + "_" + start_filename
        destination = os.path.expanduser('~') + "/Box/INFO 290T Project/Intermediate Data/Selected Images/train/" + new_file_name
        shutil.copy(source, destination)
        new_file_names.append(new_file_name)
        new_file_paths.append("~" + "/Box/INFO 290T Project/Intermediate Data/Selected Images/train/" + new_file_name)
    else:
        source = os.path.expanduser('~') + start_im_path
        new_file_name = Class + "_test_orig_" + orig_train_or_test + "_" + start_filename
        destination = os.path.expanduser('~') + "/Box/INFO 290T Project/Intermediate Data/Selected Images/test/" + new_file_name
        shutil.copy(source, destination)
        new_file_names.append(new_file_name)
        new_file_paths.append("~" + "/Box/INFO 290T Project/Intermediate Data/Selected Images/test/" + new_file_name)
# end timer
end = time.time()
print("Time to copy images in min: ", (end - start) / 60)
```

## Updated Excel file

```python
# Add new file paths to limited_cars_annos
limited_cars_annos['new_file_path'] = new_file_paths
# Add new file names to limited_cars_annos
limited_cars_annos['new_file_name'] = new_file_names
# Delete test, im_path, filename columns
limited_cars_annos = limited_cars_annos.drop(columns=['test', 'im_path', 'filename'])
# Save to "~/Box/INFO 290T Project/Intermediate Data/selected_cars_annos.xlsx"
limited_cars_annos.to_excel("~/Box/INFO 290T Project/Intermediate Data/selected_cars_annos.xlsx", index=False)
```
