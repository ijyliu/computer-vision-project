# Relabeled Train No Blur

Relabel train set items

```python
# Packages
import pandas as pd
```

## Load data

```python
# Load '../../../Data/resized_cars_annos.xlsx'
resized_cars_annos = pd.read_excel('../../../Data/resized_cars_annos.xlsx')
resized_cars_annos
```

```python
# Limit to train
resized_cars_annos_train = resized_cars_annos[resized_cars_annos['test_80_20'] == 0]
resized_cars_annos_train
```

```python
# Load and append default files in '../../../Data/Relabeled_Train_No_Blur/' for convertibles, suvs, pickups, sedans
new_labels_list = []
for label in ['convertibles', 'suvs', 'pickups', 'sedans']:
    new_labels_list.append(pd.read_csv('../../../Data/Relabeled_Train_No_Blur/' + label + '/default.txt', sep=" ", header=None))
new_labels = pd.concat(new_labels_list)
new_labels.columns = ['filename', 'label']

# In filename, remove 'No Blur/'
new_labels['filename'] = new_labels['filename'].str.replace('No Blur/', '')
# print filename for first row
print(new_labels['filename'][0])
new_labels
```

```python
# Load keys for labels '../../../Data/Relabeled_Train_No_Blur/' + label + '/synsets.txt'
# List of dataframes
class_list_dfs = []
# Convert to dataframes
for label in ['convertibles', 'suvs', 'pickups', 'sedans']:
    with open('../../../Data/Relabeled_Train_No_Blur/' + label + '/synsets.txt', 'r') as file:
        filedata = file.read()
    # Convert each line into a list item, enumerate
    class_list = filedata.split('\n')
    # Remove empty items
    class_list = list(filter(None, class_list))
    # Create dataframe of class_list with items as 'New Class' and a row number column called label
    class_list_df = pd.DataFrame(class_list, columns=['New Class'])
    class_list_df['label'] = class_list_df.index

# Check all items in class_list_dfs are equivalent
for i in range(1, len(class_list_dfs)):
    if not class_list_dfs[i].equals(class_list_dfs[0]):
        print('Not equal')
        break
    else:
        class_list_df = class_list_dfs[0]
```

# Merge data

```python
merged = pd.merge(new_labels, class_list_df, left_on='label', right_on='label', how='left')
merged
```

```python
# Old Class as filename up to first underscore
merged['Old Class'] = merged['filename'].str.split('_').str[0]
merged
```

## Create Harmonized File Name

```python
def create_harmonized_filename(file_path):
    '''
    Strip Blur vs. No Blur and other data from file path.
    '''
    # Split path by '/' and take last element
    harmonized_filename = file_path.split('/')[-1]
    # Remove _no_blur and _blurred from file name
    harmonized_filename = harmonized_filename.replace('_no_blur', '')
    harmonized_filename = harmonized_filename.replace('_blurred', '')
    # Return harmonized file name
    return harmonized_filename
```

```python
# Use funciton on 'filename' column
merged['harmonized_filename'] = merged['filename'].apply(create_harmonized_filename)
merged
```

## Basic Analytics

```python
# Share of times Old Class is not equal to New Class
print('share of times Old Class is not equal to New Class')
print(len(merged[merged['Old Class'] != merged['New Class']]) / len(merged))
```

```python
print('share of times ambiguous')
print(len(merged[merged['New Class'] == 'Ambiguous']) / len(merged))
```

## Save Data

```python
# Save file '../../../Data/Relabeled_Train_No_Blur/relabeled_train_no_blur_old_and_new_labels.xlsx'
merged.to_excel('../../../Data/Relabeled_Train_No_Blur/relabeled_train_no_blur_old_and_new_labels.xlsx', index=False)
```
