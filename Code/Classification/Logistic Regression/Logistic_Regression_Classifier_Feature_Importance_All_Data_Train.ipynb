## Load Test Data

```python
train_data = combine_directory_parquets('../../../Data/Features/All Features/train')
print('all train data')
print(train_data)

if sample_run:
    train_data = train_data.sample(frac=0.01).iloc[:, 0:10]
```

## Load Model

```python
# Load trained model
model = joblib.load('../../../Output/Classifier Fitting/' + classifier_name + '/All_Data Best Model.joblib')
model
```

## Feature Importance

```python
# Perform permutation importance
result = permutation_importance(model, X_train, y_train, n_repeats=100, random_state=222, n_jobs=-1)

# Put column name, mean and std in a dataframe
result = pd.DataFrame({'feature': train_data.select_dtypes(include=np.number).columns, 'mean': result.importances_mean, 'std': result.importances_std})

# Output to Excel
result.to_excel('../../../Output/Classifier Inference/' + classifier_name + '/train_permutation_importance.xlsx', index=False)

result
```
