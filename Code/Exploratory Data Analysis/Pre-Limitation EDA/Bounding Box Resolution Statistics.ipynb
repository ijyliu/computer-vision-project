# Resolution Statistics

```python
# Packages
import pandas as pd
import matplotlib.pyplot as plt
import dataframe_image as dfi # NOTE: YOU MUST HAVE GOOGLE CHROME INSTALLED FOR THIS TO WORK CORRECTLY
import os
```

```python
# Bring in '~\Box\INFO 290T Project\Intermediate Data\cars_annos.xslx'
cars_annos = pd.read_excel('~/Box/INFO 290T Project/Intermediate Data/cars_annos.xlsx')
cars_annos
```

```python
selected_classes_df = cars_annos[cars_annos['Class'].isin(['SUV', 'Sedan', 'Pickup', 'Coupe', 'Convertible'])]
```

```python
# Reset columns width and height
# width now x2 - x1
# height now y2 - y1
selected_classes_df['width'] = selected_classes_df['x2'] - selected_classes_df['x1']
selected_classes_df['height'] = selected_classes_df['y2'] - selected_classes_df['y1']
```

```python
# Summarize columns 'width' and 'height' by class
# percentile function
# https://stackoverflow.com/questions/17578115/pass-percentiles-to-pandas-agg-function
def percentile(n):
    def percentile_(x):
        return x.quantile(n / 100, interpolation='nearest')
    percentile_.__name__ = 'percentile_{:02.0f}'.format(n)
    return percentile_
```

```python
# Width
width_sum_stats = (selected_classes_df.groupby('Class')
                                      .agg({'width': ['count', 'mean', 'std', 'min', percentile(25), percentile(50), percentile(75), 'max']})
                                      .reset_index()
                                      .sort_values(by='Class')
                                      #.droplevel(0, axis=1)
                                      .rename(level=0, columns={'width': 'Width', 'Class': ''})
                                      .rename(columns={'count': 'Count', 'mean': 'Mean', 'std': 'Standard Deviation', 'min': 'Minimum', 'percentile_25': '25th Percentile', 'percentile_50': '50th Percentile', 'percentile_75': '75th Percentile', 'max': 'Maximum'})
                                      .style
                                      .format(precision=2, thousands=",", decimal=".")
                                      .set_table_styles([dict(selector='th', props=[('text-align', 'center')])])
                                      .set_properties(**{'text-align': 'center'})
                                      .hide(axis="index"))

# Export
dfi.export(width_sum_stats, '../../Output/Bounding Box Resolution Statistics/Bounding_Box_Width_Sum_Stats.png')

width_sum_stats
```

```python
height_sum_stats = (selected_classes_df.groupby('Class')
                                      .agg({'height': ['count', 'mean', 'std', 'min', percentile(25), percentile(50), percentile(75), 'max']})
                                      .reset_index()
                                      .sort_values(by='Class')
                                      #.droplevel(0, axis=1)
                                      .rename(level=0, columns={'height': 'Height', 'Class': ''})
                                      .rename(columns={'count': 'Count', 'mean': 'Mean', 'std': 'Standard Deviation', 'min': 'Minimum', 'percentile_25': '25th Percentile', 'percentile_50': '50th Percentile', 'percentile_75': '75th Percentile', 'max': 'Maximum'})
                                      .style
                                      .format(precision=2, thousands=",", decimal=".")
                                      .set_table_styles([dict(selector='th', props=[('text-align', 'center')])])
                                      .set_properties(**{'text-align': 'center'})
                                      .hide(axis="index"))

# Export
dfi.export(height_sum_stats, '../../Output/Bounding Box Resolution Statistics/Bounding_Box_Height_Sum_Stats.png')

height_sum_stats
```

```python
# Overall num pixels by class
selected_classes_df.groupby('Class').agg({'num_pixels': ['count', 'mean', 'std', 'min', percentile(25), percentile(50), percentile(75), 'max']}).reset_index().sort_values(by='Class')
```

```python
# Describe items in selected_classes
selected_classes_description = (selected_classes_df[['width', 'height', 'num_pixels']].rename(columns = {'width': 'Width', 'height': 'Height', 'num_pixels': 'Number of Pixels'}).describe().T
                                                                                       .rename(columns={'count': 'Count', 'mean': 'Mean', 'std': 'Standard Deviation', 'min': 'Minimum', '25%': '25th Percentile', '50%': '50th Percentile', '75%': '75th Percentile', 'max': 'Maximum'})
                                                                                       .style
                                                                                       .format(precision=2, thousands=",", decimal=".")
                                                                                       .set_table_styles([dict(selector='th', props=[('text-align', 'center')])])
                                                                                       .set_properties(**{'text-align': 'center'}))

# Export
dfi.export(selected_classes_description, '../../Output/Bounding Box Resolution Statistics/Bounding_Box_Selected_Classes_Description.png')

selected_classes_description
```
